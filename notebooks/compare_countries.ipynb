# Cross-Country Solar Potential Comparison

This notebook synthesizes cleaned solar datasets from Benin, Sierra Leone, and Togo to compare solar potential and highlight key differences across countries.

```python
# 1. Setup: Import libraries and initialize CountryComparator
import sys
sys.path.append('../')  

from src.country_comparator import CountryComparator

# Initialize the comparator with paths to cleaned data
comparator = CountryComparator(
    '../data/benin_clean.csv',
    '../data/sierraleone_clean.csv',
    '../data/togo_clean.csv'
)
```

## Metric Comparison: Boxplots of GHI, DNI, DHI by Country

```python
# Boxplots for GHI, DNI, DHI
comparator.plot_boxplots()
```

## Summary Table: Mean, Median, and Standard Deviation of GHI, DNI, DHI

```python
# Summary table for each metric by country
summary = comparator.summary_table()
print(summary)
```

## Statistical Testing: One-way ANOVA and Kruskal–Wallis Test for GHI

```python
# Statistical tests for GHI
pvals = comparator.statistical_tests()
print(f"One-way ANOVA p-value: {pvals['anova_p']:.4f}")
print(f"Kruskal–Wallis p-value: {pvals['kruskal_p']:.4f}")
```

## Key Observations
- (1) Benin shows the highest mean and median GHI, but also the greatest variability (std = 330.87), indicating both strong solar potential and high fluctuation.
- (2) Statistical tests (see p-values above) confirm that differences in GHI between countries are statistically significant, supporting meaningful cross-country comparison.
- (3) Benin is best suited for solar investment based on its higher average and median GHI, but the high variability suggests that risk management strategies should be considered; Sierra Leone has the lowest average GHI, while Togo is intermediate in both mean and variability.
